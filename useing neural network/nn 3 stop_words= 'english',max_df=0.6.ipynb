```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
from __future__ import print_function
from builtins import range
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
torch.manual_seed(142)
```

```python
# load data by using pandas command [pd.read_csv]
data= pd.read_csv('SPAM text message.csv')
data
```

```python
data['Category'].replace(['spam','ham'],[0,1],inplace=True)
```

```python
coun_vect = CountVectorizer(stop_words= 'english',max_df=0.65)
count_matrix = coun_vect.fit_transform(data.Message)
count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = coun_vect.get_feature_names())
df
```

```python
# define df as X
X=df
X
```

```python
# define target column "Category" as y 
y=data['Category']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size = 0.3,random_state=42)
```

```python
train_x = torch.Tensor(x_train.values)
test_x = torch.Tensor(x_test.values)
```

```python
#hyperparameters
hl1 = 500
hl2 = 80
#hl3 = 10
lr = 0.01
num_epoch = 800
```

```python
#build model
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(8440, hl1)
        self.fc3 = nn.Linear(hl1,hl2)
      #  self.fc4 = nn.Linear(hl2,hl2)
        #self.fc5 = nn.Linear(hl2,hl2)
        self.fc2 = nn.Linear(hl2, 2)

    def forward(self, x):
        x = F.tanh(self.fc1(x))
        x = F.tanh(self.fc3(x))
       # x = F.relu(self.fc4(x))
        #x = F.relu(self.fc5(x))
       
        x = F.sigmoid(self.fc2(x))
        return x
        
net = Net()
```

```python
#choose optimizer and loss function
criterion = nn.CrossEntropyLoss()
#optimizer = torch.optim.SGD(net.parameters(), lr=lr)
#optimizer = torch.optim.Adagrad(net.parameters(), lr=lr)
optimizer = torch.optim.Adam(net.parameters(), lr=lr)
#optimizer = torch.optim.SGD(net.parameters(), lr=lr, momentum=0.9)
#optimizer = torch.optim.RMSprop(net.parameters(), lr=lr)
```

```python
#train
import time

for epoch in range(num_epoch):
    X = torch.Tensor(train_x).float()
    Y = torch.Tensor(y_train).long()

    #feedforward - backprop
    optimizer.zero_grad()
    out = net(X)
    loss = criterion(out, Y)
    loss.backward()
    optimizer.step()
    acc = 100 * torch.sum(Y==torch.max(out.data, 1)[1]).double() / len(Y)
    if (epoch % 50 == 1):
	    print ('Epoch [%d/%d] Loss: %.4f   Acc: %.4f' 
                   %(epoch+1, num_epoch, loss.item(), acc.item()))
```

```python
"""
SECTION 3 : Testing model
"""
test_y = torch.Tensor(y_test.values)


#get prediction
X = torch.Tensor(test_x).float()
Y = torch.Tensor(test_y)
out = net(X)
_, predicted = torch.max(out.data, 1)

#get accuration
print('Accuracy of testing %.4f %%' % (100 * torch.sum(Y==predicted).double() / len(Y)))
```
